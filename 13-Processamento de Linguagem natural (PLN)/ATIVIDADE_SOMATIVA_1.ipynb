#### 1) Utilize `Expressões Regulares` para remover de cada um dos textos do corpus o cabeçalho dos capítulos (e.g., "*Capítulo 1:*", "*Capítulo 2*:", "*Capítulo 3*:")

```
import re

# Define o padrão da expressão regular para encontrar os cabeçalhos
# r'Capítulo \d+: ' -> Encontra "Capítulo ", seguido por um ou mais dígitos, um ":" e um espaço.
padrao = r'Capítulo \d+: '

# Cria uma nova lista para armazenar os textos limpos
corpus_limpo = []

# Itera sobre cada texto no corpus original
for texto in corpus:
    # Usa re.sub() para substituir o padrão encontrado por uma string vazia ""
    texto_sem_cabecalho = re.sub(padrao, '', texto)
    corpus_limpo.append(texto_sem_cabecalho)

# Imprime todos índices de 'corpus_limpo'
for linhas in corpus_limpo:
    print(linhas)
```

#### 2) Compute a quantidade de caracteres, quantidade de palavras, quantidade de palavras únicas e quantidade de sentenças de cada capítulo e apresente em um `DataFrame` da biblioteca pandas

```
import pandas as pd
import nltk
import string

nltk.download('punkt')
nltk.download('punkt_tab', quiet=True)

dados_analise = []

# Itera sobre o corpus_limpo (cabeçalho dos capítulos removidos)
for i, texto in enumerate(corpus_limpo):
    # Calcula a quantidade de caracteres
    qtd_caracteres = len(texto)

    # Calcula a quantidade de sentenças
    sentencas = nltk.sent_tokenize(texto, language='portuguese')
    qtd_sentencas = len(sentencas)

    # Tokeniza, converte para minúsculas e remove pontuação
    tokens = nltk.word_tokenize(texto.lower(), language='portuguese')
    palavras = [palavra for palavra in tokens if palavra.isalpha()]

    # Calcula a quantidade de palavras e palavras únicas
    qtd_palavras = len(palavras)
    qtd_palavras_unicas = len(set(palavras))

    # Adiciona os resultados à lista
    dados_analise.append({
        'Capítulo': i + 1,
        'Qtd. Caracteres': qtd_caracteres,
        'Qtd. Palavras': qtd_palavras,
        'Qtd. Palavras Únicas': qtd_palavras_unicas,
        'Qtd. Sentenças': qtd_sentencas
    })

# Cria o DataFrame do Pandas
df_analise = pd.DataFrame(dados_analise)

# Exibe o DataFrame formatado
print(df_analise.to_string())
```

#### 4) Imagine que durante o processo de sumarização/tradução dos textos, os dois revisores participantes referenciaram a palavra "hobbits" de maneira errada em algumas partes do texto, colocando a grafia "hobits" no lugar. Faça um programa que utilizando o algoritmo de **distância de Levenshtein**, encontre e imprima em quais capítulos são encontrados os erros de grafia.

```
from Levenshtein import distance as levenshtein_distance
import nltk

nltk.data.find('tokenizers/punkt')

# Análise com Distância de Levenshtein (usando o corpus_limpo)
palavra_correta = "hobbits"
erro_especifico = "hobits" # Definimos qual é o erro que estamos procurando
erros_encontrados = []

# Itera sobre cada texto do corpus_limpo
for i, texto in enumerate(corpus_limpo):
    # Tokeniza o texto em palavras
    palavras = nltk.word_tokenize(texto, language='portuguese')

    # Usa um conjunto para armazenar as grafias erradas únicas por capítulo
    grafias_erradas_no_capitulo = set()

    for palavra in palavras:
        # Calcula a distância de Levenshtein, ignorando case sensitive
        palavra_lower = palavra.lower()
        dist = levenshtein_distance(palavra_lower, palavra_correta)

        # A distância deve ser 1 E a palavra deve ser exatamente o erro que procuramos.
        if dist == 1 and palavra_lower == erro_especifico:
            grafias_erradas_no_capitulo.add(palavra)

    # Se encontramos erros, armazena o resultado
    if grafias_erradas_no_capitulo:
        erros_encontrados.append({
            "Capítulo": i + 1,
            "Grafias Incorretas": list(grafias_erradas_no_capitulo)
        })

#Apresentação dos Resultados
if erros_encontrados:
    print("Erros de grafia para a palavra 'hobbits' foram encontrados nos seguintes capítulos:\n")
    for erro in erros_encontrados:
        print(f"- Capítulo {erro['Capítulo']}: Encontrada(s) a(s) palavra(s) -> {', '.join(erro['Grafias Incorretas'])}")
else:
    print("Nenhum erro de grafia para a palavra 'hobbits' foi encontrado no corpus.")
```

#### 5) Remova as stop-words do corpus e mostre os vetores resultantes (com suas palavras correspondentes) do processo de BoW.
> DICA: Você pode utilizar um atributo da própria função [CountVectorizer](https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html)



```
import pandas as pd
import nltk
from sklearn.feature_extraction.text import CountVectorizer

nltk.data.find('tokenizers/punkt')
# Adiciona o download para 'stopwords'
print("Baixando o pacote 'stopwords' do NLTK...")
nltk.download('stopwords')


# Remoção de Stop Words e Vetorização com BoW

# Carrega a lista de stop words em português
stopwords_pt = nltk.corpus.stopwords.words('portuguese')

# Instancia o CountVectorizer, passando a lista de stop words
# Ele também irá tokenizar e converter para minúsculas por padrão
vectorizer = CountVectorizer(stop_words=stopwords_pt)

# Aplica o vetorizador ao corpus limpo para criar a matriz termo-documento
X = vectorizer.fit_transform(corpus_limpo)

# Recupera os nomes das palavras (features) que formam o vocabulário
palavras_vocabulario = vectorizer.get_feature_names_out()

# Cria os nomes para as colunas (Capítulos)
nomes_colunas = [f'Capítulo {i+1}' for i in range(len(corpus_limpo))]

# Cria o DataFrame com a matriz
df_bow = pd.DataFrame(X.T.toarray(), index=palavras_vocabulario, columns=nomes_colunas)

# Apresentação dos Resultados
print("Vetorização Bag of Words (BoW) com Remoção de Stop Words\n")
print("Valores representam a frequência de cada palavra por capítulo.\n")

# Exibe o DataFrame completo
# O .to_string() garante que todas as linhas e colunas sejam mostradas
print(df_bow.to_string())
```

#### 6) Faça um algoritmo que leia uma palavra do usuário e em seguida mostre o capítulo mais relevante a pesquisa, juntamente ao correspondente valor de pontuação TF-IDF da palavra buscada no documento.

```
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.data.find('corpora/stopwords')

# Pré-processamento
stopwords_pt = nltk.corpus.stopwords.words('portuguese')

# Cálculo da Matriz TF-IDF
vectorizer = TfidfVectorizer(stop_words=stopwords_pt)
X = vectorizer.fit_transform(corpus_limpo)
palavras_vocabulario = vectorizer.get_feature_names_out()
nomes_colunas = [f'Capítulo {i+1}' for i in range(len(corpus_limpo))]
df_tfidf = pd.DataFrame(X.T.toarray(), index=palavras_vocabulario, columns=nomes_colunas)


# Interação com o Usuário e Análise
# Solicita a palavra ao usuário
palavra_busca = input("Digite a palavra que deseja pesquisar no corpus: ").lower()
print("-" * 50)

# Verifica se a palavra existe no vocabulário gerado
if palavra_busca in df_tfidf.index:
    # Pega a linha (vetor de pontuações) correspondente à palavra
    pontuacoes = df_tfidf.loc[palavra_busca]

    # Encontra o capítulo com a maior pontuação TF-IDF
    capitulo_mais_relevante = pontuacoes.idxmax()
    pontuacao_maxima = pontuacoes.max()

    # --- Apresentação dos Resultados ---

    # 1. Mostra o capítulo mais relevante
    if pontuacao_maxima > 0:
        print(f"A palavra '{palavra_busca}' é mais relevante no '{capitulo_mais_relevante}'.")
        print(f"Pontuação TF-IDF nesse capítulo: {pontuacao_maxima:.4f}\n")
    else:
        # Isso pode acontecer se a palavra for uma stop word ou não estiver no corpus
        print(f"A palavra '{palavra_busca}' foi encontrada, mas não é considerada relevante em nenhum capítulo (pontuação TF-IDF é 0).\n")

    # 2. Mostra o DataFrame com a pontuação em todos os capítulos
    print(f"Pontuação TF-IDF para a palavra '{palavra_busca}' em cada capítulo:")

    # Cria um DataFrame apenas com a linha da palavra buscada para melhor visualização
    df_palavra = pd.DataFrame(pontuacoes).T
    print(df_palavra.to_string())

else:
    print(f"A palavra '{palavra_busca}' não foi encontrada no vocabulário do corpus (após remoção de stop words).")



```
