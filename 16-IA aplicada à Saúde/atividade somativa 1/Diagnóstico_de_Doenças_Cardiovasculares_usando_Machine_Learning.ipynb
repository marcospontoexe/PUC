Antes de iniciar a análise dos dados, precisamos compreender o significado de cada variável que compõe o *dataset*. Assim, abaixo, podemos visualizar a lista com a descrição de todos os atributos.

**Informações sobre os atributos:**


* `age` - idade em anos
* `sex` - sexo do paciente
  * 0: mulher
  * 1: homem
* `cp` - tipo da dor torácica
  * 1: angina típica
  * 2: angina atípica
  * 3: dor não cardíaca
  * 4: assintomática
* `trestbps` - pressão arterial em repouso
* `chol` - colesterol sérico (*mg/dl*)
* `fbs` - açucar no sangue em jejum > 120*mg/dl*
  * 0: `False`
  * 1: `True`
* `restecg` - resultado do eletrocardiografia de repouso
  * 0: normal
  * 1: anormalidades de ST-T  (inversão da onda T e elevação ou depressão de > 0.05*mV*)
  * 2: hipertrofia ventricular esquerda provável ou definitiva (pelos critérios de Romhilt-Estes)
* `thalach` - frequência cardíaca máxima atingida
* `exang` - angina induzida pelo exercício
  * 0: não
  * 1: sim
* `oldpeak` - depessão do segmento ST induzida pelo exercício em relação ao repouso
* `slope` - inclinação do segmento ST no pico do exercício
* `ca` - número de vasos principais colorido por fluoroscopia
* `thal` - *teste de esforço cardíaco*
  * 3: normal
  * 6: defeito fixo
  * 7: defeito reversível
* `num` - diagnóstico de doença cardíaca (diagnóstico angiográfico)
  * 0: não há doença cardíaca ( < 50% de estreitamento do diâmetro)
  * 1,2,3,4: há doença cardíaca ( > 50% de estreitamento do diâmetro)

### **Preparação dos Dados**

**Nesta etapa do projeto serão utilizados os dados do *dataset* df_uci**

Em todos os projetos de Machine Learning a etapa de pré-processamento de dados é fundamental. Aplicando as técnicas corretas para preparação dos dados, podemos aumentar consideravelmente o desempenho dos modelos implementados.

Para alimentar o modelo de Machine Learning, é necessário separar a variável alvo (dependente) do *dataset* das demais (independentes).

Geralmente, as variáveis são divididas como X e y, este projeto vai adotar o mesmo padrão.

Após separar nossas variáveis devidamente, como parte do pré-processamento dos dados, que irão alimentar o modelo de Machine Learning, será usado o **StardardScaler**, que vem junto com **sklearn.preprocessing**, para padronizar nossos dados numéricos.

```python
# DIVISÃO DOS DADOS EM TREINO E TESTE

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_uci.drop(['num'], axis=1),   # remove as colunas label 
    df_uci['num'],   # Coluna que será usada como label de classificação
    test_size=0.2,  # informamos a porcentagem de divisão para a base de testes.
    random_state=35  # aqui informamos um "seed".
)

```

```python
# IMPORTAÇÃO DAS FERRAMENTAS PARA O PRÉ-PROCESSAMENTO DOS DADOS
from sklearn.preprocessing import StandardScaler

# PADRONIZAÇÃO DOS DADOS NUMÉRICOS

# Vamos usar o StandardScaler, que padroniza os dados (média=0, desvio padrão=1).
scaler = StandardScaler()

X_train_array = scaler.fit_transform(X_train)
X_test_array = scaler.transform(X_test)

# Convertendo de volta para um DataFrame
X_train_scaler = pd.DataFrame(X_train_array, columns=X_train.columns, index=X_train.index)
X_test_scaler = pd.DataFrame(X_test_array, columns=X_test.columns, index=X_test.index)
```

Por fim, antes de implementar o modelo, iremos dividir o nosso *dataset* em dados de treino e de teste.

Essa etapa também é fundamental, uma vez que ela evita que o modelo possa fazer previsões erradas ao "viciar", em palavras simples, o modelo com um determinado conjunto de dados. Ou seja, quando nós usamos os mesmos dados para treinar e testar nossos modelos, é provável que ele não aprenda a classificar os dados, mas apenas "decorar" as respostas. Assim, ao submetir esse modelo para novos dados, ele não vai conseguir uma boa acurácia.

Por isso, é fundamental usar dados diferentes para treino e teste.

### **Modelo Random Forest**

Random Forest (floresta aleatória) é um algoritmo de aprendizagem supervisionada. Como você pode perceber pelo seu nome, ele cria uma floresta de um modo aleatório.

Segundo Niklas Donges, em um artigo do Towards Data Science, de modo simples: o algoritmo de florestas aleatórias cria várias árvores de decisão e as combina para obter uma predição com maior acurácia e mais estável.

<center><img alt = "Analisando Airbnb" width = "15%" src="https://static.thenounproject.com/png/961660-200.png"></center>

Uma grande vantagem do algoritmo de florestas aleatórias é que ele pode ser utilizado tanto para tarefas de classificação quanto para regressão, o que representa a maioria dos sistemas de aprendizagem de máquina atuais.

Você pode conferir um artigo completo do **Towards Data Science** sobre o modelo Random Forest clicando [aqui](https://towardsdatascience.com/understanding-random-forest-58381e0602d2).

```python
# IMPORTANDO O MODELO RANDOM FOREST

from sklearn.ensemble import RandomForestClassifier
```

```python
# INSTANCIANDO O MODELO

RF_model = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 35)
```

```python
# TREINANDO O MODELO

RF_model.fit(X_train_scaler, y_train)
```

### **Verificando o desempenho do modelo**

Uma das principais métricas utilizadas para avaliar o desempenho de um modelo de Machine Learning é a acurácia. Contudo, é importante ressaltar que cada caso é particular, com isso, uma boa acurácia nem sempre garantirá que o modelo está bom.

No caso da detecção de doenças cardiovasculares, o interesse é maximizar o número de verdadeiros positivos e minimizar o número de verdadeiros positivos e minimizar o número de falsos negativos, apesar de ser uma tarefa difícil de realizar.

Neste projeto, além da métrica de acurácia será utilizada a função classification_report para ver ver o desempenho do modelo sobre as métricas de precisão, ***recall, f1-score e support***.

```python
# IMPORTAÇÃO DAS FERRAMENTAS DE MÉTRICAS

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# REALIZANDO A PREVISÃO NOS DADOS DE TESTE

y_pred = RF_model.predict(X_test_scaler)
```

```python
# VIZUALIZANDO A ACURÁCIA GERAL

print('[Acurácia] Random Forest:', accuracy_score(y_test, y_pred))
```

```python
# VISUALIZANDO O CLASSIFICATION REPORT

print('\n[Classification Report] Random Forest')

print(classification_report(y_test, y_pred))
```

**Descrição das métricas**

* **Precision** - Precisão de previsões positivas..
* **Reacall** - Fração de positivos que foram identificados corretamente.
* **f1-score** - representa a média harmonica entre precision e recall, sendo 0 o pior valor e 1 o melhor.

Aparentemente, o modelo apresenta bons resultados nas métricas avaliadas. Contudo, como citado anteriormente, isso não pode garantir totalmente que o modelo seja bom.

Para complementar a nossa análise de desempenho, é interessante visualizar a **matriz de confusão**

<center><img alt = "Analisando Airbnb" width = "50%" src="https://miro.medium.com/max/938/1*s7VB26Cfo1LdVZcLou-e0g.png"></center>

* **VP** - Se o paciente tiver doença cardíaca e o modelo prever que possui.
* **FP** - Se o paciente não tiver doença cardíaca e o modelo prever que possui.
* **FN** - Se o paciente tiver doença cardíaca e o modelo prever que não.
* **VN** - Se o paciente não tiver doença cardíaca e o modelo prever que não.

```python
# VISUALIZANDO A MATRIZ DE CONFUSÃO

pd.DataFrame(confusion_matrix(y_test, y_pred),
             index=['neg', 'pos'], columns=['pred_neg', 'pred_pos'])
```

A matriz de confusão funciona da seguinte maneira:

Nas linhas constam os casos reais e nas colunas, as previsões feitas pelo modelo.

Por exemplo, existem 35 casos realmente negativos e o modelo previu corretamente. Contudo, em 7 casos também negativos, ele previu como positivo.

### **Conclusões - Modelo de Machine Learning**

## Variáveis numéricas

Variáveis como `age`, `trestbps`, `chol` e `thalach` são numéricas. Para elas, as principais estatísticas são as de **tendência central** e as de **dispersão**.

| Estatística | O que é? | Por que é Importante na Análise?  |
| :--- | :--- | :--- |
| **Contagem (Count)** | Número total de registros não nulos. | É o primeiro passo para verificar se há **dados faltantes**. Se a contagem for menor que o total de linhas, sabemos que precisamos tratar esses dados ausentes. |
| **Média (Mean)** | A soma de todos os valores dividida pela contagem. | Dá uma noção do **valor típico** da variável. Ex: "A idade média dos pacientes na amostra é de 54 anos". A média é sensível a valores extremos (*outliers*). |

Se a mediana da pressão arterial (`trestbps`) for 130 e a média for 145, isso sugere a presença de alguns pacientes com pressão muito alta puxando a média para cima. |

Um desvio baixo indica que eles se concentram perto da média. Ex: Um alto desvio padrão no colesterol (`chol`) mostra uma grande diversidade de perfis de pacientes, desde saudáveis até de alto risco. **Resumo para Variáveis Numéricas:** Uma tabela com `count`, `mean`, `std`, `min`, `25%`, `50%` (mediana), `75%`, `max` é o ponto de partida padrão para qualquer análise exploratória.

###  Correlação (Entre Variáveis Numéricas)

Uma etapa adicional, mas ainda básica, é calcular a **matriz de correlação**.

*   **O que é?** Mede a força e a direção da relação linear entre pares de variáveis numéricas. O valor varia de -1 a +1.
    *   **+1:** Correlação positiva perfeita (quando uma sobe, a outra sobe).
    *   **-1:** Correlação negativa perfeita (quando uma sobe, a outra desce).
    *   **0:** Nenhuma correlação linear.
*   **Por que é importante?**
    *   **Entender Relações:** Ajuda a validar hipóteses. Por exemplo, poderíamos esperar uma correlação negativa entre `thalach` (frequência cardíaca máxima) e `age` (idade), já que a frequência máxima tende a diminuir com a idade.
    *   **Identificar Redundância:** Variáveis com correlação muito alta (ex: > 0.9) podem ser redundantes, e talvez uma delas possa ser removida para simplificar o modelo (*feature selection*).
    *   **Classes alvo**: Classes que tem alta correlação com a label a ser classificada é muito importante para o modelo de predição, já as que tem baixa correlação na maioria das vezes não são tão importantes e podem ser descartadas.

Essas estatísticas, juntas, fornecem um panorama completo e robusto do seu conjunto de dados, sendo um passo indispensável antes de qualquer visualização ou modelagem.

    *  **outliers** (valores atípicos), que aparecem como pontos fora dos "bigodes" do gráfico.
    *  **dispersão dos dados** (o tamanho da caixa)
    *  **simetria dos dados**, quando a linha da mediana está no centro(Q2). Se a mediana esta mais para o alto (Q3), existe
        uma distribuição assimétrica negativa, e vice-versa. 
* Exemplo: Um box plot para chol (colesterol) provavelmente mostraria aquele valor de 564 mg/dl como um outlier claro, muito acima do resto dos pacientes.

# 3. Usando modelos preditivos
Como teste serão testados 4 modelos:
1.  Regressão Logística: calcula a probabilidade de um paciente pertencer à classe "doente" com base em uma combinação linear das variáveis de entrada.
2.  SVM: Tenta encontrar a "linha" ou "hiperplano" que melhor separa as duas classes (pacientes doentes e não doentes) no espaço de dados, com a maior margem possível entre elas.
3.  XGBoost: Também usa árvores, mas de forma sequencial. Ele constrói uma árvore, vê onde ela errou, e constrói a próxima árvore para corrigir os erros da anterior. É um processo de aprendizado iterativo e focado nos erros.

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import VarianceThreshold
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

## Variance Threshold + Scaler + Regressão Logística
Este é o nosso modelo de base: simples, rápido e interpretável.

```python
# vamos verificar qual é a variancia de cada coluna
print("\nVariância de cada coluna do dataset de treino:")
print(X_train.var())
print("-" * 30)
```

```python
# Definir o pipeline
pipeline_lr = Pipeline([
    ('variance_threshold', VarianceThreshold(threshold=0.1)), # Remove features com variância menor que 0.1
    ('scaler', scaler), # Padroniza os dados
    ('logistic_regression', LogisticRegression(random_state=35, max_iter=1000)) # O modelo
])

#  Treinar o pipeline
pipeline_lr.fit(X_train, y_train)

#  Fazer previsões e avaliar
y_pred_lr = pipeline_lr.predict(X_test)
accuracy_lr = accuracy_score(y_test, y_pred_lr)

print(f"Acurácia: {accuracy_lr:.4f}")
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred_lr, target_names=['Sem Doença', 'Com Doença']))
```

## Variance Threshold + Scaler + SVM (Support Vector Machine)
SVM é um modelo poderoso que busca a melhor "fronteira" para separar as classes

```python
# --- Pipeline 2: SVM ---

#  Definir o pipeline
pipeline_svm = Pipeline([
    ('variance_threshold', VarianceThreshold(threshold=0.1)),
    ('scaler', scaler), # Essencial para o SVM
    ('svm', SVC(kernel='rbf', probability=True, random_state=35)) # 'rbf' é um bom kernel padrão
])

#  Treinar o pipeline
pipeline_svm.fit(X_train, y_train)

# 3. Fazer previsões e avaliar
y_pred_svm = pipeline_svm.predict(X_test)
accuracy_svm = accuracy_score(y_test, y_pred_svm)

print(f"Acurácia: {accuracy_svm:.4f}")
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred_svm, target_names=['Sem Doença', 'Com Doença']))
```

## Variance Threshold + Scaler + XGBoost
Frequentemente o modelo com a melhor performance para dados tabulares.

```python
# --- Pipeline 3: XGBoost ---

#  Definir o pipeline
pipeline_xgb = Pipeline([
    ('variance_threshold', VarianceThreshold(threshold=0.1)),
    ('scaler', scaler),
    ('xgboost', XGBClassifier(eval_metric='logloss', random_state=35))
])

# 2. Treinar o pipeline
pipeline_xgb.fit(X_train, y_train)

# 3. Fazer previsões e avaliar
y_pred_xgb = pipeline_xgb.predict(X_test)
accuracy_xgb = accuracy_score(y_test, y_pred_xgb)

print(f"Acurácia: {accuracy_xgb:.4f}")
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred_xgb, target_names=['Sem Doença', 'Com Doença']))
```

```python
# Criar um DataFrame com os resultados
results = pd.DataFrame({
    'Modelo': ['Regressão Logística', 'SVM', 'XGBoost'],
    'Acurácia': [accuracy_lr, accuracy_svm, accuracy_xgb]
})

# Ordenar os resultados
results = results.sort_values(by='Acurácia', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(
    x='Acurácia', 
    y='Modelo', 
    data=results, 
    palette='viridis', 
    hue='Modelo',      # <<-- PASSO 1: Atribua a variável do eixo Y ao 'hue'
    legend=False       # <<-- PASSO 2: Desative a legenda redundante
)
plt.title('Comparação de Acurácia dos Pipelines', fontsize=16)
plt.xlabel('Acurácia', fontsize=12)
plt.ylabel('Modelo', fontsize=12)
plt.xlim(0.7, 1.0)
plt.show()

print("\nRanking de Performance dos Modelos:")
print(results)
```

Isso que quase é muito importante para treinamento de um modelo preditivo, pois se o dataset tivesse por exemplo apenas 10% de labels positivas, ficaria quase impossível fazer uma previsão verdadeiro-positivo.
* Apenas 32% das features são do sexo feminino, isso pode tornar a previsão para o sexo feminino mais imprecisa doque para o sexo masculino.
* XGBoost, por serem modelos de conjunto, tendem a ter uma ligeira vantagem em termos de acurácia, embora a diferença possa não ser drástica devido à qualidade dos dados.
