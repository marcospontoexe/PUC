# Atividade somativa 1

## Visão Geral do Dataset


Começaremos montando uma sequencia básica;
1. Carregar o dataset.
2. Limpeza de dataset.
   * Remoção de atributos correlacionados.
   * Remoção de instâncias vazias ou duplicadas.
3. Divisão do dataset para dados de treino e de teste.
4. Preparação dos dados:
   * Remoção de outliers.
   * seleção de atributos.
   * Padronização.
5. Treinamento do modelo.
6. Analise de resultado.
7. Conclusões finais.
